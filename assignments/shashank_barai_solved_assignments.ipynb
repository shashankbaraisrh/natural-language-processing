(1) Create a Python program that reads a text file and prints all the email
addresses found in the file using regular expressions. Extend the program to
extract and print domain names of the email addresses found.

```python
#re module is used for Regular expressions to perform various text manipulation tasks such as searching, finding, and replacing patterns within strings.

import re

def find_emails_and_domains(file_path):
    # Regular expression for matching email addresses
    email_regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

    # Open the file and read its contents
    with open(file_path, 'r') as file:
        file_content = file.read()

        # Find all email addresses using the regular expression
        emails = re.findall(email_regex, file_content)

        # Extract domain names from email addresses
        domain_names = []
        for email in emails:
            domain = email.split('@')[1]
            domain_names.append(domain)

        # Print the found email addresses and domain names
        print("Email addresses found in the file:")
        for email, domain in zip(emails, domain_names):
            print(f"Email: {email}, Domain: {domain}")

# I used ADSA whatsap group and scraped all its text into one file called whatsap_messages.text 
file_path = r'C:\Users\User\Desktop\shashankbaraicollege\NLP\whatsapp_messages.txt'
  
find_emails_and_domains(file_path)
```

(2) Write a Python program that reads a text file containing a list of phone
numbers in various formats (e.g., (123) 456-7890, 123-456-7890,
123.456.7890) and converts them to a standardized format (e.g., 1234567890)
using regular expressions. Handle country codes and different variations of
phone number formats.

```python
#re module is used for Regular expressions to perform various text manipulation tasks such as searching, finding, and replacing patterns within strings.

import re

def standard_phone_numbers(file_path):
    # Regular expression for matching phone numbers in various formats
    phone_regex = r'\(?\d{3}\)?[\s.-]?\d{3}[\s.-]?\d{4}'

    # Open the file and read its contents
    with open(file_path, 'r') as file:
        file_content = file.read()

        # Finding all phone numbers using the regular expression
        phone_numbers = re.findall(phone_regex, file_content)

        # Standardize and print the phone numbers
        print("Standard phone numbers:")
        for phone_number in phone_numbers:
            standardized_number = re.sub(r'\D', '', phone_number)  # Remove non-digit characters
            print(standardized_number)

#I used my university applied data science course whatsap group and scraped all its text into one file called whatsap_messages.text
file_path = r'C:\Users\User\Desktop\shashankbaraicollege\NLP\whatsapp_messages.txt' 
standard_phone_numbers(file_path)
```

(3) Develop a Python script that extracts all the URLs from a given HTML
document using regular expressions.

```python
#re module is used for Regular expressions to perform various text manipulation tasks such as searching, finding, and replacing patterns within strings.

import re

def extract_urls_from_html_document(html_content):
    # Regular expression for matching URLs in HTML
    url_regex = r'https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+'

    # Find all URLs using the regular expression
    urls = re.findall(url_regex, html_content)

    # Print the extracted URLs
    print("Extracted URLs:")
    for url in urls:
        print(url)

# Example usage
html_content = """
<html>
<head>
<title>Example</title>
</head>
<body>
<p>This is an HTML document with some URLs:</p>
<a href="https://www.google.com">Google</a>
<a href="https://www.wikipedia.org">Wikipedia</a>
<a href="https://www.github.com">GitHub</a>
</body>
</html>
"""

extract_urls_from_html_document(html_content)
```

(4) Write a Python program that extracts all the hashtags from a given tweet
using regular expressions.

```python
#re module is used for Regular expressions to perform various text manipulation tasks such as searching, finding, and replacing patterns within strings.
import re

def extract_hashtags(tweet):
    # Regular expression for matching hashtags
    hashtag_regex = r'\B#\w*[a-zA-Z]+\w*'

    # Find all hashtags using the regular expression
    hashtags = re.findall(hashtag_regex, tweet)

    # Print the extracted hashtags
    print("Extracted hashtags:")
    for hashtag in hashtags:
        print(hashtag)

# Example usage
tweet = "One day I will become great #AI Specialist #Data_Scientist! #Gen_AI #inspiration"
extract_hashtags(tweet)
```

(6) Develop an interactive Python application using Tkinter or PyQt that allows
users to input text and view the stemmed or lemmatized output along with
the original words.

```python
# Importing necessary libraries i.e NLTK for natural language processing tasks like tokenization, stemming, and lemmatization. It also utilizes the Counter class for counting word occurrences. 
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Downloading NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(text):
    # Tokenize the text in to words
    tokens = word_tokenize(text)

    # Remove punctuation and stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.isalnum() and word.lower() not in stop_words]

    return filtered_tokens

def apply_stemming(tokens):
    # Apply stemming
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]

    return stemmed_tokens

def apply_lemmatization(tokens):
    # Apply lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    return lemmatized_tokens

# My text
text = "In my first semester I and my team Developed the Hushhush Recruiter AI model using supervised and unsupervised machine learning algorithms, alongside frontend development with HTML, CSS, and JavaScript."

# Preprocessing the text
tokens = preprocess_text(text)

# Applying stemming
stemmed_tokens = apply_stemming(tokens)

# Applying lemmatization
lemmatized_tokens = apply_lemmatization(tokens)

print("Original Text:", tokens)
print("Stemmed Text:", stemmed_tokens)
print("Lemmatized Text:", lemmatized_tokens)
```

(7) Write a Python script that continuously monitors a text stream (e.g., news
articles from RSS feeds) and performs stemming and lemmatization on new
incoming texts in real-time. Display the processed text along with some
statistics (e.g., most common stems or lemmas).

```python
import time# used to introduce delays or pauses between fetching new batches of text from the RSS feed.
import feedparser #feedparser allows the script to fetch and parse the RSS feed from a website.
from nltk.stem import PorterStemmer#Porter stemming algorithm is used for stemming words
from nltk.tokenize import word_tokenize#To split a string of text into individual words or tokens
from collections import Counter#The Counter class is a dict subclass used to count the occurrences of elements in a list

# Initialize stemmer
stemmer = PorterStemmer()

# Function to process text and display statistics
def process_and_display(text):
    tokens = word_tokenize(text)  # Tokenize the text
    stems = [stemmer.stem(token) for token in tokens]  # Stem each token
    stem_counts = Counter(stems)  # Count the occurrences of each stem
    print("Most common stems:")
    for stem, count in stem_counts.most_common(5):
        print(f"{stem}: {count}")

# Function to continuously monitor the RSS feed
def monitor_rss_feed(url):
    while True:
        feed = feedparser.parse(url)  # Fetch the latest feed
        for entry in feed.entries:
            text = entry.title + " " + entry.summary  # Combining title and summary
            print("\nOriginal Text:")
            print(text)
            process_and_display(text)  # Process and display the text
        time.sleep(60) 

# Main function
def main():
    
    rss_feed_url = "https://rss.nytimes.com/services/xml/rss/nyt/World.xml"  
    monitor_rss_feed(rss_feed_url)  #Monitoring the RSS feed

if __name__ == "__main__":
    main()
```
